```python
%matplotlib inline
import numpy as np 
import matplotlib.pyplot as plt 
```

```python

class Board:
     def __init__(self):
        self.grid = np.ones((8, 8, 3), np.float64)
        div = 1
        for colum in range(8):
            for row in range(8):
                div = 0 if div else 1
                if row % 2 == div and colum % 2 == div:
                    self.grid[colum][row] = np.array((0, 0, 0))
  

     def add_red(self, row, colum):
        red = (1, 0, 0)
        self.grid[row][colum] = red
        self.red = [row, colum]

     def add_blue(self, row, colum):
        blue = (0, 0, 1)
        self.grid[row][colum] = blue
        self.blue = [row, colum]

     def render(self):
        plt.imshow(self.grid)

     def is_under_attack(self):
        
        if self.blue[0]==self.red[0] or self.blue[1]==self.red[1]:
            return True 
        if abs(self.blue[0]-self.red[0])==abs(self.blue[1]-self.red[1]):#The abs() function is used to return the absolute value of a number.
            return True
               
        return False
```

```python
# Diagonal True
b = Board()
b.add_blue(0, 2)
b.add_red(2, 4)
assert b.is_under_attack() == True
print(b.is_under_attack())
b.render()
```

```python

# Same row true
b = Board()
b.add_blue(2, 2)
b.add_red(2, 6)
print(b.is_under_attack())
b.render()
```

```python
# Same Column True
b = Board()
b.add_blue(3, 4)
b.add_red(5, 4)
print(b.is_under_attack())
b.render()
```
